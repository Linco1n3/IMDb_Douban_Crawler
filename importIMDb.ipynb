```python
links = pd.read_csv("links.csv")
links.info()
```

```python
movieIDs = links["imdbId"]
print(movieIDs[0])
print(len(str(movieIDs[0])))
```

```python
# id, title, year, language, description, production companies, genres, rating
with open('IMDb.csv', mode='w') as f:
    fieldnames = ['id','title', 'year', 'language', 'plot outline', 'production companies', 'genres', 'rating', 'director']
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    count=0
    for movieid in movieIDs:
        try:
            movieid = str(movieid).zfill(7)
            movie = ia.get_movie(movieid)
            row = [int(movieid), 
                    movie.data["title"], 
                    movie.data["year"],
                    movie.data["languages"][0],
                    movie.data["plot outline"],
                    movie.data["production companies"][0],
                    movie.data["genres"][0],
                    movie.data["rating"],
                    movie["director"][0]['name']]
            dict_writer = dict(zip(fieldnames, row))
            #print(dict_writer)
            count+=1
            if count%100 == 0:
                print("start processing movie number: ", count)
            writer.writerow(dict_writer)
        except Exception as e:
            continue
```

```python
imdb = pd.read_csv("IMDb.csv")
keep_col = ['id','title','year','language','plot outline', 'production companies', 'genres', 'rating']
new_imdb = imdb[keep_col]
new_imdb.to_csv("new_IMDb.csv", index=False)
```
