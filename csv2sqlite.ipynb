```python
import pandas as pd
import sqlite3
from pandas.io import sql
import subprocess
 
# define input .csv file and output .sqlite file
in_csv = './DoubanTop250.csv'
out_sqlite = './IMDb.sqlite'
 
table_name = 'DoubanTop250' # define table name
chunksize = 5000 # define chunksize, not necessary but speed up large transcation
 
# column name
columns = ['rank','ChineseName','originalName','rating','votes']
 
# number of rows
nlines = subprocess.check_output(['wc', '-l', in_csv])
nlines = int(nlines.split()[0])
 
# connect to database
cnx = sqlite3.connect(out_sqlite)
 
# main loop
for i in range(1, nlines, chunksize):  # drop row=0 which is header of csv file
 
    df = pd.read_csv(in_csv,
            header=None,
            nrows=chunksize,
            skiprows=i)
 
    df.columns = columns
 
    sql.to_sql(df,
                name=table_name,
                con=cnx,
                index=False,
                index_label='rank',
                if_exists='append')
cnx.close()
```
