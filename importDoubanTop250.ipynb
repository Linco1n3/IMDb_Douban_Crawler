```python
import re
import requests
from bs4 import BeautifulSoup
from openpyxl import Workbook
import csv
```

```python
header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 UBrowser/6.1.2107.204 Safari/537.36'}
wb = Workbook()
ws = wb.active
for page in range(0,250,25):
    url='https://movie.douban.com/top250?start=' + str(page) + '&filter='
    url_get = requests.get(url,headers = header)
    url_soup = BeautifulSoup(url_get.text,'html.parser')
    li = url_soup.find('ol',attrs={"class":"grid_view"}).find_all('li')
    for subject in li:
        ranking = re.findall('<em class="">(.*?)</em>',str(subject))
        name = re.findall('<span class="title">(.*?)</span>',str(subject)) # actually it provides more than one name
        rating = re.findall('property="v:average">(.*?)</span>',str(subject))
        votes = re.findall('<span>(.*?)人评价</span>',str(subject))
        ws['A' + ranking[0]] = ranking[0]
        ws['B' + ranking[0]] = name[0] # get its Chinese Name
        ws['C' + ranking[0]] = name[-1].replace('/','').lstrip().rstrip() # return its original name and remove redundant space
        ws['D' + ranking[0]] = rating[0]
        ws['E' + ranking[0]] = votes[0]
        print(ranking[0],name[0],name[-1].replace('/','').lstrip().rstrip())
```

```python
wb.save('top250.xlsx')
```
